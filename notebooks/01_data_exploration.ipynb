```python
# Cell 1: Import Required Libraries

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Make plots appear inside notebook
%matplotlib inline

# Optional: cleaner plot style
sns.set(style="whitegrid")
```

```python
# Cell 2: Load Dataset (from raw folder)

df = pd.read_csv("../data/raw/uci-ml-phishing-dataset.csv")

df.head()

```

```python
# Cell 3: Dataset Shape and Info

print("Shape of dataset:", df.shape)

df.info()
```

```python
# Cell 4: Missing Values

df.isnull().sum()
```

```python
# Cell 5: Class Distribution

df['Result'].value_counts()
```

```python
# Cell 6: Convert Labels

df['Result'] = df['Result'].map({-1: 0, 1: 1})

df['Result'].value_counts()
```

```python
# Cell 7: Correlation Heatmap

plt.figure(figsize=(12, 10))
sns.heatmap(df.corr(), cmap="coolwarm")
plt.title("Feature Correlation Matrix")
# SAVE BEFORE SHOW
plt.savefig("../results/figures/Feature_Correlation_Matrix.png", dpi=300, bbox_inches="tight")

plt.show()
```
